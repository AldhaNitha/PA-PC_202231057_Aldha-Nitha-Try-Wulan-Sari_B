```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar asli
image = cv2.imread('Aldha.jpg')

# Fungsi untuk menampilkan gambar-gambar dalam satu baris
def show_images(images, titles):
    fig, axes = plt.subplots(1, len(images), figsize=(20, 10))
    for idx, ax in enumerate(axes):
        ax.imshow(cv2.cvtColor(images[idx], cv2.COLOR_BGR2RGB))
        ax.set_title(titles[idx])
        ax.axis('off')
    plt.show()

# Transformasi gambar
# 1. Gambar Asli
original_image = image.copy()

# 2. Rotasi
(height, width) = image.shape[:2]
center = (width // 2, height // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

# 3. Perubahan Ukuran
resized_image = cv2.resize(image, (width // 2, height // 2))

# 4. Pemotongan
start_x, start_y = width // 4, height // 4
end_x, end_y = start_x + width // 4, start_y + height // 4
cropped_image = image[start_y:end_y, start_x:end_x]

# 5. Pembalikan
flipped_image = cv2.flip(image, 1)

# 6. Translasi
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
translated_image = cv2.warpAffine(image, translation_matrix, (width, height))

# Menampilkan semua gambar
images_list = [original_image, rotated_image, resized_image, cropped_image, flipped_image, translated_image]
titles_list = ['Gambar Asli', 'Setelah Rotasi', 'Setelah Resize', 'Setelah Crop', 'Setelah Flip', 'Setelah Translasi']

show_images(images_list, titles_list)
```
